```python
#Import dependencies
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
import re
from sqlalchemy import create_engine
```

```python
# Importing csv files to create dataframe
movie_data = "data/movies.csv"
movies_df = pd.read_csv(movie_data)
movies_df
```

```python
# Importing csv files to create dataframe
ratings_data = "data/ratings.csv"
ratings_df = pd.read_csv(ratings_data)
ratings_df["timestamp"]=pd.to_datetime(ratings_df["timestamp"],unit='s')
ratings_df
```

```python
# Importing tags csv files to create dataframe
tags_data = "data/tags.csv"
tags_df = pd.read_csv(tags_data)
tags_df["timestamp"]=pd.to_datetime(tags_df["timestamp"],unit='s')
tags_df
```

## Combining the three datasets to a single dataset 

```python
user_movies_ratings_combined_df=pd.merge(movies_df,ratings_df,on="movieId",how="inner")
user_movies_ratings_combined_df.rename(columns={"timestamp":"rating_timestamp"},inplace=True)
user_movies_ratings_combined_df
```

```python
movies_ratings_tags_combined_df=pd.merge(user_movies_ratings_combined_df,tags_df,on=["movieId","userId"],how="left")
movies_ratings_tags_combined_df.rename(columns={"timestamp":"tags_timestamp"},inplace=True)
movies_ratings_tags_combined_df
```

Data Transformation

```python
movie_avg_rating=ratings_df.groupby("movieId").agg({"rating":"mean"}).reset_index()
```

```python
movie_avg_rating.head()
```

```python
movies_df=movies_df.merge(movie_avg_rating,on="movieId",how="left")
```

```python
def find_year(input_string):
    try:
        x=re.search(r"\(\d\d\d\d\)",input_string)
        year=str(x.group(0))
        year=year.replace('(','')
        year=year.replace(')','')
        return year
    except:
        return "-1"
```

```python
#Creating column with year of movie
movies_df['year'] = movies_df['title'].apply(find_year)
movies_df['year'].unique()
#movies_df[movies_df['year'].isna()]
#movies_df=movies_df.fillna(-1)
```

```python
movies_df.rename(columns={"rating":"avg_user_rating"},inplace=True)
```

```python
movies_df["avg_user_rating"]=movies_df["avg_user_rating"].fillna(-1)
```

```python
movies_df["genres"]=movies_df["genres"].str.replace('|',' ')
movies_df
```

```python
#renaming (no genres listed) to 
movies_df = movies_df.replace('(no genres listed)', 'UnKnown')
movies_df
```

```python
# Creating a column combining all the features
movies_df['comb']=movies_df['genres']+' '+movies_df['avg_user_rating'].round(4).map(str)+' '+movies_df['year'].map(str)
#movies_df['genres']+' '+' '+str(movies_df['year'])+' '+str(movies_df['avg_user_rating'])
```
